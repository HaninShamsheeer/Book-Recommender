Use text classification to work with the categories column

```python
import pandas as pd
books = pd.read_csv("books_cleaned.csv")
books["categories"].value_counts().reset_index()
```

```python
## Take a deeper look into some of the categories
books[books["categories"] == "Juvenile Fiction"]
```

```python
## We may combine these specific categories as fiction vs. non fiction
books["categories"]
```

Mapping the top 10 categories as either Fiction, Nonfiction, Children's Fiction, Children's Nonfiction


```python
category_mapping = {
    "Fiction" : "Fiction",
    "Juvenile Fiction" : "Children's Fiction",
    "Biography & Autobiography" : "Nonfiction",
    "History" : "Nonfiction",
    "Literary Criticism" : "Nonfiction",
    "Philosophy" : "Nonfiction",
    "Religion" : "Nonfiction",
    "Comics & Graphic Novels" : "Fiction",
    "Drama" : "Fiction",
    "Juvenile Nonfiction" : "Children's Nonfiction",
    "Science" : "Nonfiction",
    "Poetry" : "Nonfiction"
}

## Making a new simple column that maps top 10 categories to 4 specific categories
books['simple_category'] = books['categories'].map(category_mapping)
books
```

```python
books.simple_category.isna().value_counts()
```

We can see that 3743 of the total books have a simple category and 1454 of the books are without a simple category


```python
## Use Zero-Shot Classification from HuggingFace to classify categories into simple categories
from transformers import pipeline

fiction_categories = ["Fiction", "Nonfiction"]

pipe = pipeline("zero-shot-classification",
                model="facebook/bart-large-mnli",
                device="cpu")
```

```python
## Show descriptions of simple category = fiction
books.loc[books["simple_category"] == "Fiction", "description"].reset_index(drop=True)[0]

## Make a sequence of text that we will pass through the zero-shot classifier
sequence = books.loc[books["simple_category"] == "Fiction", "description"].reset_index(drop=True)[0]
```

```python
## Passing through the classifier
pipe(sequence, fiction_categories)
```

This gives us a score of the probability of whether it is fiction or not. In this particular case, probability of Fiction is much higher


```python
## We try to extract the label with the higher probability
import numpy as np
max_index = np.argmax(pipe(sequence, fiction_categories)["scores"]) ## gives us the score of the higher probability
max_label = pipe(sequence, fiction_categories)["labels"][max_index] ## gives us the label of the higher score
max_label
```

```python
## We now generalize and create a function that does the same
def generate_predictions(sequence, categories):
    predictions = pipe(sequence, categories)
    max_index = np.argmax(predictions["scores"])
    max_label = predictions["labels"][max_index]
    return max_label

actual_categories = []
predicted_categories = []
```

```python
## Test the model

from tqdm import tqdm

## For the fiction category
for i in tqdm(range(0, 300)):
    sequence = books.loc[books["simple_category"] == "Fiction", "description"].reset_index(drop=True)[i]
    predicted_categories += [generate_predictions(sequence, fiction_categories)]
    actual_categories += ["Fiction"]
```

```python
## For the Nonfiction category
for i in tqdm(range(0, 300)):
    sequence = books.loc[books["simple_category"] == "Nonfiction", "description"].reset_index(drop=True)[i]
    predicted_categories += [generate_predictions(sequence, fiction_categories)]
    actual_categories += ["Nonfiction"]
```

```python
## Put both the actual and predicted categories into a dataframe to compare them
predictions_df = pd.DataFrame({"actual_cat": actual_categories, "predicted_cat": predicted_categories})
predictions_df
```

Pretty close!


```python
## To see how many of the predictions were correct
predictions_df["correct_predictions"] = (
    np.where(predictions_df["actual_cat"] == predictions_df["predicted_cat"], 1, 0)
)
correct_percent = predictions_df.correct_predictions.sum()/len(predictions_df)
correct_percent
```

We see that the 77% of the predictions matched the actual predictions

```python
## Now we may use this model to predict the categories of the missing categories
## First we extract the description and isbn of books with missing categories
isbns = []
predicted_cats = []

missing_cats = books.loc[books.simple_category.isna(), ["isbn13", "description"]].reset_index(drop=True)
missing_cats
```

```python
## Then we try to predict the categories of these missing books using the above model that we constructed

for i in tqdm(range(0, len(missing_cats))):
    sequence = missing_cats.loc[i, "description"]
    predicted_label = generate_predictions(sequence, fiction_categories)
    predicted_cats.append(predicted_label)
    isbns.append(missing_cats.loc[i, "isbn13"])

## This populates predicated_cats with the predictions of categories and isbns with its corresponding isbn
```

```python
## Testing the two lists

print(isbns[0:10],len(isbns))
```

```python
## Testing the two lists

print(predicted_cats[0:10],len(predicted_cats))
```

Looks like both, isbns and predicted_cat works like we want it to

```python
## We make a dataframe with the predicted categories and their isbns
missing_predicted_df = pd.DataFrame({"isbn13": isbns, "predicted_categories": predicted_cats})
missing_predicted_df
```

```python
## We now merge this missing dataframe to the original dataframe, which is easier because we have the isbns of the books in the missing dataframe

books = pd.merge(books, missing_predicted_df, on="isbn13", how="left")
books
```

```python
## Now we amend the simple categories to so that if they are empty, they take the values from the predicted category

books.simple_category = np.where(books.simple_category.isna(), books.predicted_categories, books.simple_category)
books
```

We see that the simple category is now fully populated, there are no empty values there.

```python
## We may now drop the predicted categories column since we do not have any use for it anymore

books = books.drop(columns = ["predicted_categories"])
books
```

Now we have successfully merged the predicted categories of the missing categories to the simple categories and dropped the extra column

```python
## We now save the books with predicted categories as a new file
books.to_csv("books_with_categories.csv", index=False)
```
