```python
import kagglehub
from seaborn import heatmap

# Download latest version
path = kagglehub.dataset_download("dylanjcastillo/7k-books-with-metadata")
print("Path to dataset files:", path)

# This gives us the location of the kaggle dataset that we will be working with --> which we can read into pandas
```

```python
import pandas as pd
```

```python
# Read the data at the path into books
books = pd.read_csv(f"{path}/books.csv")
```

```python
books
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
## Analysis of Missing Values
ax = plt.axes()
sns.heatmap(books.isnull().transpose(), cbar = False, ax = ax)
plt.xlabel("Columns")
plt.ylabel("Missing Values")
```

```python
## Making two new columns (missing description and age of the book)
import numpy as np
books['missing_description'] = np.where(books.description.isnull(), 1, 0)
books['age_of_book'] = 2026 - books.published_year
```

```python
## Extracting columns of interest
columns_of_interest = ['num_pages', 'age_of_book', 'missing_description','average_rating']

## Making a correlation matrix of columns of interest: we are trying to figure out if the missing description has any correlation with the some of the columns
correlation_matrix = books[columns_of_interest].corr(method="spearman") ## Method is spearman since missing description is a non-continuous var

sns.set_theme(style="white")
plt.figure(figsize=(8,6))
heatmap = sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap="coolwarm", cbar_kws={"label": "Spearman Correlation"})
heatmap.set_title("Correlation Heatmap")
plt.show()
```

There is not any significant correlation coefficients since most are closer to 0 than to 1, we may assume that it might be relatively safe to delete the books that do not have any observations

```python
## Extracting books that have missing description, pages, average rating, published year
book_missing = books[(books.description.isna()) |
    (books.num_pages.isna()) |
    (books.average_rating.isna()) |
    (books.published_year.isna())]
```

This gives us 303 values which given the size of our data set (7,000) is only ~4% of books. We many remove these

```python
## Books with complete values
books_complete = books[~(books.description.isna()) &
    ~(books.num_pages.isna()) &
    ~(books.average_rating.isna()) &
    ~(books.published_year.isna())]
books_complete
```

This is the dataset with all books with complete information

Looking at distinct categories, the number of categories is ~500+ which is unusable in its current quantity

```python
## Looking at the distribution of categories in the descending order of number of books in each category
books_complete["categories"].value_counts().reset_index().sort_values("count", ascending = False)
```

What this graph tells us: a lot of the categories are concentrated in a few of the categories. We must normalize them in order to effictively use them in our model

```python
## We will try to remove those descriptions that are not useful
books_complete['words_in_description'] = books_complete['description'].str.split().str.len()
books_complete
```

This is the distribution of the number of words in the description, we can create an arbitrary bound to the number of words required

```python
## Bins

## Descriptions with 1 to 4 words
books_complete.loc[books_complete['words_in_description'].between(1, 4), 'description']
```

As we can see, descriptions with 1-4 words are not really useful

```python
## Descriptions with 5 to 14 words
books_complete.loc[books_complete['words_in_description'].between(5, 14), 'description']
```

Not enough to be useful

```python
## Descriptions with 15 to 24 words
books_complete.loc[books_complete['words_in_description'].between(15, 24), 'description']
```

Again, too short to be useful

```python
## Descriptions with 25 to 34 words
books_complete.loc[books_complete['words_in_description'].between(25, 34), 'description']
```

Now these can be useful. We can use words above 25 as the bound for our description

```python
## New dataframe with more than 25 words in the description
books_complete_25_words = books_complete[books_complete['words_in_description'] >= 25]
books_complete_25_words
```

There are 5197 unique books in the dataframe now which is pretty good to be the basis of our recommender

```python
## Join the title and the subtitle into one if subtitle exists
books_complete_25_words["title_and_subtitle"] = (
    np.where(books_complete_25_words["subtitle"].isna(), books_complete_25_words["title"],
             books_complete_25_words[["title", "subtitle"]].astype(str).agg(": ".join, axis = 1)))
books_complete_25_words
```

```python
## In order to make a vector database, we create a new column with the isbn and the description
books_complete_25_words['tagged_description'] = books_complete_25_words[["isbn13", "description"]].astype(str).agg(" ".join, axis = 1)
books_complete_25_words
```

We have finished the cleaning phase of the project, we may now drop columns that we may not need and save it for further use

```python
## Drop some columns we dont need and save it to a csv file

( books_complete_25_words
  .drop(["subtitle", "missing_description", "age_of_book", "words_in_description"], axis = 1)
  .to_csv("books_cleaned.csv", index = False) )

pd.read_csv("books_cleaned.csv")
```
